## 3. Create the Necessary Datasets of Stars ## 

This section goes through the steps of creating the datapoints for the constellations. 

1. View the given x,y,z coordinates. 

```python
x = [-0.41, 0.57, 0.07, 0.00, -0.29, -0.32,-0.50,-0.23, -0.23]
y = [4.12, 7.71, 2.36, 9.10, 13.35, 8.13, 7.19, 13.25,13.43]
z = [2.06, 0.84, 1.56, 2.07, 2.36, 1.72, 0.66, 1.25,1.38]
```

## 4. Perform Visualizations of the Constellations ##

This section goes over the visualizations of the stars. 

1. Create a 2D map of the stars using the scatter plot. 

```python
# Create a figure
fig = plt.figure(figsize=(10,10))

# Add your subplot
ax = fig.add_subplot(1,1,1)

# Create a scatter plot
plt.scatter(x,y)
plt.title('2D Positions of the Stars')
plt.grid(True, alpha = 0.5)
plt.show()
```

**Visual Summary:** The scatter plot above shows the 2D position of the stars in the constellations. 

2. Create a 3D visualization of the Orion Constellation with the (x,y,z) dataset coordinates. 

```python
# Create a figure
fig_3d = plt.figure(figsize=(5,5))
ax = fig_3d.add_subplot(1, 1, 1, projection="3d")

# Create a scatter plot
plt.scatter(x,y,z)

plt.show()
```

3. Create a 2D constellations plot with night-time theme. 

```python
# Create a figure
fig = plt.figure(figsize=(10,10))
# Add your subplot
ax = fig.add_subplot(1,1,1)
# Change color of plot background
ax.set_facecolor((0.0, 0.0, 0.0))

# Create a scatter plot with black background
ax.scatter(x,y,color=(1.0, 1.0, 1.0))
plt.show()
```

4. Create a night-time themed 3D constellation of Orion. 

```python
# Create a figure
fig_3d = plt.figure(figsize=(5,5))
# Add your subplot
ax = fig_3d.add_subplot(1, 1, 1, projection="3d")
# Change color of plot background
ax.set_facecolor((0.05, 0.0, 0.1))

# Create a scatter plot with black background
ax.scatter(x,y,z,color=(1.0,1.0,1.0))
plt.show()
```
